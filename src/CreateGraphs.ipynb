```python
import matplotlib.pyplot as plt
import pandas as pd
from utils import buildRunName
import os
from typing import Any

matplotlibColors = plt.rcParams["axes.prop_cycle"].by_key()["color"]
```

```python
def extractRunNameFromJson(fileName: str) -> str:
    fileName = fileName.removeprefix("logs_fit_")
    fileName = fileName.removesuffix("_train.json")
    fileName = fileName.removesuffix("_validation.json")
    return fileName
```

```python
accuracyBasePath = "../graphs/data/accuracy/"
jsonLogFilesAccuracy = os.listdir(accuracyBasePath)
falsePositivesBasePath = "../graphs/data/false_positives/"
jsonLogFilesFalsePositives = os.listdir(falsePositivesBasePath)
falseNegativesBasePath = "../graphs/data/false_negatives/"
jsonLogFilesFalseNegatives = os.listdir(falseNegativesBasePath)
transferAccuracyBasePath = "../graphs/data/transfer/accuracy/"
jsonLogFilesTransferAccuracy = os.listdir(transferAccuracyBasePath)
batchBasePath = "../graphs/data/batch/"
jsonLogFilesBatch = os.listdir(batchBasePath)
```

```python
accuracyTrain = dict()
accuracyValidation = dict()
for jsonLogFile in jsonLogFilesAccuracy:
    if jsonLogFile.__contains__("train"):
        accuracyTrain[extractRunNameFromJson(jsonLogFile)] = pd.read_json(
            accuracyBasePath + jsonLogFile
        )
    elif jsonLogFile.__contains__("validation"):
        accuracyValidation[extractRunNameFromJson(jsonLogFile)] = pd.read_json(
            accuracyBasePath + jsonLogFile
        )
```

```python
falsePositivesTrain = dict()
falsePositivesValidation = dict()
for jsonLogFile in jsonLogFilesFalsePositives:
    if jsonLogFile.__contains__("train"):
        falsePositivesTrain[extractRunNameFromJson(jsonLogFile)] = pd.read_json(
            falsePositivesBasePath + jsonLogFile
        )
    elif jsonLogFile.__contains__("validation"):
        falsePositivesValidation[extractRunNameFromJson(jsonLogFile)] = pd.read_json(
            falsePositivesBasePath + jsonLogFile
        )
```

```python
falseNegativesTrain = dict()
falseNegativesValidation = dict()
for jsonLogFile in jsonLogFilesFalseNegatives:
    if jsonLogFile.__contains__("train"):
        falseNegativesTrain[extractRunNameFromJson(jsonLogFile)] = pd.read_json(
            falseNegativesBasePath + jsonLogFile
        )
    elif jsonLogFile.__contains__("validation"):
        falseNegativesValidation[extractRunNameFromJson(jsonLogFile)] = pd.read_json(
            falseNegativesBasePath + jsonLogFile
        )
```
